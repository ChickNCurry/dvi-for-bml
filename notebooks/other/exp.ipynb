```python
sin = 0.5 * np.sin(np.linspace(0, 2 * np.pi, 50))
sin_points = np.random.choice(sin.shape[0], size=3)

sin2 = np.sin(3 * np.linspace(0, 2 * np.pi, 50))
sin2_points = np.random.choice(sin2.shape[0], size=4)

sin3 = 0.7 * np.sin(1.5 * np.linspace(0, 2 * np.pi, 50))
sin3_points = np.random.choice(sin3.shape[0], size=3)

plt.figure(figsize=(2, 1))
plt.scatter(sin_points, sin[sin_points], c="r", marker="X", zorder = 2)
plt.plot(sin, zorder = 1)
plt.ylim(-1, 1)
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.axis("off")
plt.show()

plt.figure(figsize=(2, 1))
plt.scatter(sin2_points, sin2[sin2_points], c="r", marker="X", zorder = 2)
plt.plot(sin2, zorder = 1)
plt.ylim(-1, 1)
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.axis("off")
plt.show()

plt.figure(figsize=(2, 1))
plt.scatter(sin3_points, sin3[sin3_points], c="r", marker="X", zorder = 2)
plt.plot(sin3, zorder = 1)
plt.ylim(-1, 1)
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.axis("off")
plt.show()
```

```python
def plot(func, x, x_c, x_t):
    plt.figure(figsize=(4.5, 1.5))

    plt.scatter(x_c, func(x_c), c="r", marker="X", zorder = 2)
    plt.scatter(x_t, func(x_t), c="black", marker="X", zorder = 2)
    plt.plot(x, func(x), zorder = 1)
    
    plt.ylim(-1.2, 1.2)
    plt.yticks([-1, -0.5, 0, 0.5, 1])
    plt.axis("off")
    plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 100)
x_c = np.random.choice(x, size=3)
x_t = np.random.choice([i for i in x if i not in x_c], size=6) # 
func = lambda x: 0.7 * np.sin(x)

plot(func, x, x_c, x_t)

x = np.linspace(0, 2 * np.pi, 100)
x_c = np.random.choice(x, size=3)
x_t = np.random.choice([i for i in x if i not in x_c], size=6)
func = lambda x: 0.5 * np.sin(3 * x)

plot(func, x, x_c, x_t)

x = np.linspace(0, 2 * np.pi, 100)
x_c = np.random.choice(x, size=3)
x_t = np.random.choice([i for i in x if i not in x_c], size=6)
func = lambda x: np.sin(6 * x)

plot(func, x, x_c, x_t)
```

```python
context_size = 5
z_dim = 2
num_samples = 10000

context = torch.rand((num_samples, context_size, z_dim), device=device) * 3
# context = torch.ones((context_size, z_dim), device=device) * 1.5
print(context.shape)

p_z = ContextualLatentSpaceGMM(context)

print(p_z.mu.shape)
print(p_z.sigma.shape)
print([g.mean for g in p_z.gaussians])
```

```python
samples = p_z.sample()

print(samples.shape)

plt.hist2d([z[0] for z in samples], [z[1] for z in samples], bins=40, range=[(-5, 5), (-5, 5)])
plt.ylim(-5, 5)
plt.xlim(-5, 5)
plt.show()
```

```python
sin = 0.5 * np.sin(np.linspace(0, 2 * np.pi, 50))
sin_points = np.random.choice(sin.shape[0], size=3)

sin2 = np.sin(3 * np.linspace(0, 2 * np.pi, 50))
sin2_points = np.random.choice(sin2.shape[0], size=4)

sin3 = 0.7 * np.sin(1.5 * np.linspace(0, 2 * np.pi, 50))
sin3_points = np.random.choice(sin3.shape[0], size=3)

plt.figure(figsize=(6, 3))
plt.scatter(sin_points, sin[sin_points], c="r", marker="X", zorder = 2)
plt.plot(sin, zorder = 1)
plt.ylim(-1, 1)
plt.yticks([-1, -0.5, 0, 0.5, 1])
# plt.axis("off")
plt.show()

plt.figure(figsize=(6, 3))
plt.scatter(sin2_points, sin2[sin2_points], c="r", marker="X", zorder = 2)
plt.plot(sin2, zorder = 1)
plt.ylim(-1, 1)
plt.yticks([-1, -0.5, 0, 0.5, 1])
# plt.axis("off")
plt.show()

plt.figure(figsize=(6, 3))
plt.scatter(sin3_points, sin3[sin3_points], c="r", marker="X", zorder = 2)
plt.plot(sin3, zorder = 1)
plt.ylim(-1, 1)
plt.yticks([-1, -0.5, 0, 0.5, 1])
# plt.axis("off")
plt.show()
```
